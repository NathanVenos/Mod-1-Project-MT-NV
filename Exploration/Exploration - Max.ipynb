```python
#checking duplicate rows of tconst
duplicates = imbd_title_basics[imbd_title_basics.duplicated('tconst')]
print(len(duplicates))
duplicates.head()
```

```python
#creating imbd dataframe
imbd_df = imbd_title_basics
```

```python
#replacing empty genre cells with "Missing"
imbd_df['genres'].replace(np.nan, 'Missing', inplace=True, regex=True)
```

```python
#setting index column for imbd_df
imbd_df.set_index("tconst", inplace=True)
```

```python
#creating imbd_ratings dataframe
imbd_ratings_df = imbd_title_ratings
```

```python
#setting index column for imbd_ratings
imbd_ratings_df.set_index("tconst", inplace=True)
```

```python
#joining imbd_df with imbd_ratings
imbd_df_join1 = imbd_df.join(imbd_ratings_df, how='left')
imbd_df_join1.head()
```

```python
#rename tconst column
imbd_title_akas = imbd_title_akas.rename(columns={'title_id': 'tconst'})
imbd_title_akas.head()
```

```python
#setting index column for imbd_df
imbd_title_akas.set_index("tconst", inplace=True)
imbd_title_akas.head()
```

```python
#joining imbd_df with imbd_title_akas
imbd_df_join2 = imbd_df_join1.join(imbd_title_akas, how='left')
imbd_df_join2.head()
```
